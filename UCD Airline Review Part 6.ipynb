```python
import matplotlib.pyplot as plt

labels = 'Food_&_Beverages', 'Inflight_Entertainment', 'Seat_Comfort', 'Staff_Service', 'Value_for_Money'
sizes = [3, 2, 3, 4, 3]
explode = (0, 0, 0, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
Above Stats show in percentage what Criteria Aer Lingus do Best in
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()

x=[1, 2, 3, 4, 5]
y=[3, 2, 3, 4, 3]

y2=[2, 2, 2, 2, 2]

ax.bar(x,y,color= "magenta")
ax.bar(x,y2,color= "green")

plt.show()
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()

x=[1, 2, 3, 4, 5]
y=[3, 2, 3, 4, 3]
y2=[2, 1, 2, 3, 3]

ax.scatter(x,y,color= "magenta")
ax.plot(x,y2,color= "green", marker="*")

plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Aer Lingus", "British Airways", "Etihad Airways", "Emirates", "Ryanair", "United Airlines", "Jet Blue", "Air Asia", "Easy Jet"]
w = [6, 5, 4, 6, 5, 3, 5, 5, 5]
c = ["red", "green", "orange", "blue", "magenta", "black", "pink"]
plt.barh(x,w,0.5,color=c)
plt.xlabel("Rating")
plt.ylabel("Airline")
plt.title("Airline Rating out of 10")
plt.show()
```

```python
Seen Above is a Sample rating Score out of 10 for 7 Airlines using a Horizontal Bar Chart.
```
